```python
import yfinance as yf
import pandas as pd

# Download BTC-USD data till today
df = yf.download(
    "BTC-USD",
    start="2019-01-01",
    interval="1d",
    progress=False
)

# Reset index so Date becomes a column
df.reset_index(inplace=True)

# Save to CSV
df.to_csv("BTC_USD_Yahoo.csv", index=False)

print("CSV downloaded successfully")
print(df.tail())
```
